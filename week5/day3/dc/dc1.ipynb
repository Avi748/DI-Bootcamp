```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 100).astype("int")
y = np.square(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(title='Plot of x vs y (y=x^2)',xlabel='x', ylabel='y')
# fig.savefig('fig.png');
```
