```python
#7
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig_1 = plt.figure(1, figsize=(8.4, 4.8))

chart_1 = fig_1.add_subplot(121)
chart_2 = fig_1.add_subplot(122)

chart_1.plot(x, y1, 'tab:blue')
chart_2.scatter(x, y2, linestyle='--');
```

```python
#9
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(x, y1, label='y = sin(x)')
ax2.scatter(x, y2, label='y = cos(x)')
ax1.legend()
ax2.legend();
```
