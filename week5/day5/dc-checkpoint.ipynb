### Visualize the Data:

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
df['Sex'].value_counts().plot(kind='pie')
plt.title('Ratio of male to female passengers');
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', kde=True, bins=30)
plt.title('Distribution of Age');
```

```python
s = df['Survived'].replace([0, 1], ["didnt survive", 'survived'])

plt.figure(figsize=(10, 6))
s.value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.title('Number of passengers who survived and didnt survive');
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex');
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Correlation between Age and Ticket Fare');
```

### Preprocess the Data:

```python
df['Age'].fillna(df['Age'].mean())
```

### Data Analysis:

```python
#survival rate for different genders
diff_gender = pd.crosstab(s,df['Sex'])
diff_gender
```

```python
#survival rate for different passenger classes
diff_class = pd.crosstab(s,df['Pclass'])
diff_class
```
